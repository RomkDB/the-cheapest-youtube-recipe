## 전처리 방식 고안
1. [SAUCEPIA] 가 들어가면 '[SAUCEPIA]' 글자와 앞뒤 공백 삭제함
2. 괄호가 들어가면, (괄호)만 삭제함
3. 끝에 숫자, 영어가 들어가면 숫자, 영어만 삭제함
4. 특수문자가 들어가면 삭제함
5. 최종 중복 제거

```python
df2 = df.copy()

# 데이터 전처리 함수 정의
def preprocess_title(title):
    # [SAUCEPIA]가 포함되어 있으면 제거하고 앞뒤 공백 제거
    if '[SAUCEPIA]' in title:
        title = title.replace('[SAUCEPIA]', '').strip()
    # 괄호와 괄호 안의 문자 제거
    title = re.sub(r'\([^)]*\)', '', title)
    # 숫자, 영어 제거
    title = re.sub(r'[0-9a-zA-Z]', '', title)
    # 특수문자 제거
    title = re.sub(r'[^\w\s]', '', title)
    # 공백 제거
    title = title.strip()
    return title

# 데이터프레임에 전처리 함수 적용
df2['title'] = df2['title'].apply(preprocess_title)

# 중복 제거
df2.drop_duplicates(subset='title', keep='first', inplace=True)

print(df2)
print("총 menu : ", len(df2))
```

```python
# CSV 파일로 저장
df2.to_csv('../3_데이터전처리/sauce_recipe_2365.csv', index=False)
```
