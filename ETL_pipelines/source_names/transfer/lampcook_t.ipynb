### 메뉴 컬럼에 괄호 처리가 제대로 되어있지 않아서 다시 처리함

```python
filtered_df = df[df['Menu'].str.contains('\)')]
print(filtered_df)
```

```python
import re

# 메뉴 컬럼의 내용을 수정하는 함수
def modify_menu(menu):
    # 괄호가 포함된 경우
    if '(' in menu:
        # 첫 번째 여는 괄호의 위치 찾기
        first_open_bracket_index = menu.find('(')
        # 첫 번째 닫는 괄호의 위치 찾기
        first_close_bracket_index = menu.find(')')
        
        # 첫 번째 닫는 괄호가 있는 경우
        if first_close_bracket_index != -1:
            # 괄호와 그 내용 삭제
            menu = menu[:first_open_bracket_index] + menu[first_close_bracket_index + 1:]
        else:
            # 첫 번째 닫는 괄호가 없는 경우, 괄호 이후 내용 모두 삭제
            menu = menu[:first_open_bracket_index]

    # 문자열 끝에 있는 ')' 삭제
    menu = re.sub(r'\)$', '', menu)
    
    return menu.strip()  # 양쪽 공백 제거 후 반환


# 메뉴 컬럼 내용 수정 적용
df['Menu'] = df['Menu'].apply(modify_menu)
```

## 전처리 방식 고안
1. 이상치 제거
2. 괄호가 들어가면, '(괄호)' 내용을 삭제함
3. 끝에 영어가 들어가면 영어만 삭제함
4. 특수문자가 들어가면 삭제함
5. 최종 중복 제거

```python
df2 = df.copy()

# 이상치 제거
outlier = [6054, 6057, 6058, 6059, 6060, 6061, 6063, 6064, 6065, 6073, 6074, 6754, 7905, 7955, 7956, 7957, 7958, 8679]
df2 = df2.drop(outlier)

# Menu 컬럼의 내용에 괄호가 들어가면 (괄호)만 삭제
#df2['Menu'] = df2['Menu'].str.replace(r'\([^\)]*\)', '')

# Menu 컬럼의 내용 끝에 영어가 들어가면 영어만 삭제
#df2['Menu'] = df2['Menu'].str.replace(r'[a-zA-Z]+$', '')
df2['Menu'] = df2['Menu'].apply(lambda x: re.sub(r'[a-zA-Z]+$', '', x))

# Menu 컬럼의 내용에 특수문자가 들어가면 해당 행 삭제
df2 = df2[~df2['Menu'].str.contains(r'[^\w\s]')]

# 최종 Menu 내용 기준으로 중복 제거
df2 = df2.drop_duplicates(subset=['Menu'])

print(df2)
print("총 menu : ", len(df2))
```

```python
# 확인용
# 정규표현식 패턴 설정
pattern = re.compile(r'[a-zA-Z]+$')

# Menu 컬럼에서 정규표현식 패턴과 일치하는 값을 가지는 행 추출
rows_with_special_chars = df2[df2['Menu'].str.contains(pattern, na=False)]

print(rows_with_special_chars)
```
