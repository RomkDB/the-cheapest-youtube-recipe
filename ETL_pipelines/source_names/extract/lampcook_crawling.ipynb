```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import pandas as pd
import re  # 정규표현식 모듈 추가

# WebDriver 생성
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

try:
    # 각 메뉴별 페이지 수, 카테고리
    page_nums = [232, 20, 22, 31, 159]
    categories = ['한식', '중식', '일식', '양식', '퓨전 및 기타']

    # 빈 리스트 생성
    menu_data = []

    for menu_num, page_num in enumerate(page_nums, start=1):
        # 각 메뉴별 페이지에 접속
        url = f'http://www.lampcook.com/food/food_dic_menu{menu_num}.php?search_mode=0&alpha_no=0&sql_one=&pagenum=1'
        driver.get(url)
        
        # 각 페이지의 span_bold 텍스트 (메뉴명) 추출
        for page in range(1, page_num + 1):
            # 페이지 이동
            driver.get(f'http://www.lampcook.com/food/food_dic_menu{menu_num}.php?search_mode=0&alpha_no=0&sql_one=&pagenum={page}')

            # BeautifulSoup을 사용하여 HTML 파싱
            soup = BeautifulSoup(driver.page_source, 'html.parser')

            # span_bold 텍스트 추출 및 괄호와 괄호 안의 영어 삭제
            span_bold_elements = soup.find_all('div', class_='span_bold')
            for element in span_bold_elements:
                text = element.text.strip()
                
                # 괄호와 그 안의 영어 부분 삭제 및 괄호 안의 내용 추출
                cleaned_text = re.sub(r'\((?![ㄱ-힣])[^)]*\)', '', text)
                parentheses_text = re.findall(r'\((?![ㄱ-힣])[^)]*\)', text)

                # 메뉴 데이터 저장
                menu_data.append({'Menu': cleaned_text, 'Category': categories[menu_num - 1], '괄호': parentheses_text})

                print(categories[menu_num - 1], page, cleaned_text, parentheses_text)

    # DataFrame 생성
    df = pd.DataFrame(menu_data)

    # DataFrame을 CSV 파일로 저장
    csv_filename = 'lampcook.csv'
    df.to_csv(csv_filename, index=False)

    print(f"CSV 파일 '{csv_filename}'로 저장되었습니다.")

finally:
    # WebDriver 종료
    driver.quit()
```
