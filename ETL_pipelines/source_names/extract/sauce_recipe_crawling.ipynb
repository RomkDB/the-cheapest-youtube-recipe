```python
import requests
from bs4 import BeautifulSoup

# 크롤링할 웹 페이지 URL
# 설명 : 링크 에러가 있어서 모바일에서 유효한 링크 가져옴
# 주의 : 페이지 옮길 시 원래 링크로 돌아감 (~recipe.do) -> 따라서 계속 여기로 접속하여 페이지 이동 시도
url = 'https://sauce.foodpolis.kr/home/recipe/recipe.do?PAGE_MN_ID=SIS-020100'

# 웹 페이지에 HTTP GET 요청 보내고 응답 받기 (SSL 인증서 에러를 방지하기 위해 verify False 설정)
response = requests.get(url, verify = False)

# HTTP 요청이 성공했는지 확인
if response.status_code == 200:
    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 웹 페이지에서 원하는 요소를 찾아내고 데이터 추출
    no_elements = soup.find_all('td', class_='m')
    title_elements = soup.find_all('td', class_='text-left')

    for title_element in title_elements:
        a_tag = title_element.find('a')
        if a_tag:
            title = a_tag.text.strip()
            print(title)

    # 추가적으로 필요한 작업 수행 가능
else:
    print('HTTP 요청 실패:', response.status_code)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import pandas as pd

# WebDriver 생성
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

try:
    # 크롤링할 첫 페이지 URL
    url = 'https://sauce.foodpolis.kr/home/recipe/recipe.do?PAGE_MN_ID=SIS-020100'

    # 첫 페이지 접속
    driver.get(url)

    # 빈 리스트 생성
    no_list = []
    title_list = []
    category_list = []

    while True:
        # BeautifulSoup을 사용하여 HTML 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # 웹 페이지에서 원하는 요소를 찾아내고 데이터 추출
        no_elements = soup.find_all('td', class_='m')
        title_elements = soup.find_all('td', class_='text-left')
        category_elements = soup.select('tr td:last-child')

        for title_element in title_elements :
            a_tag = title_element.find('a')
            if a_tag:
                title = a_tag.text.strip()
                title_list.append(title)
        
        for no_element, category_element in zip(no_elements, category_elements):
            no = no_element.text.strip()
            no_list.append(no)
            category = category_element.text.strip()
            category_list.append(category)
            print(no)
        
        if no == '1' :
            break

        # '다음 페이지' 버튼 탐색
        try:
            next_button = driver.find_element(By.XPATH, '//li[@class="active page-item"]/following-sibling::li[@class="page-item"]')
        except:
            next_button = driver.find_element(By.XPATH, '//li[@class="next"]/a')

        # "다음" 버튼 클릭
        if next_button:
            next_button.click()
            time.sleep(2)  # 페이지가 로드될 때까지 기다림
        else:
            print('더 이상 다음 페이지가 없습니다.')
            break

    # DataFrame 생성
    df = pd.DataFrame({'no': no_list, 'title': title_list, 'category': category_list})

    # DataFrame을 CSV 파일로 저장
    csv_filename = 'sauce_recipe_menu.csv'
    df.to_csv(csv_filename, index=False)

    print(f"CSV 파일 '{csv_filename}'로 저장되었습니다.")

finally:
    # WebDriver 종료
    driver.quit()
```
